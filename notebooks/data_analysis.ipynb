```python
from collections import Counter
import pandas as pd
import ipywidgets as widgets
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
from wordcloud import WordCloud, STOPWORDS
sns.set_theme()
warnings.filterwarnings("ignore")
```

**Tag Distribution**

```python
df = pd.read_csv('/home/viethung/MLOps-Fundamentals/datasets/labeled_projects.csv')
```

```python
# Distribution of tags
tags = df['tag'].to_list()

# Create a dictionary to store the frequency of each element
frequency = {}

# Loop through the list and count the frequency of each element
for item in tags:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

values = frequency.values()
tags = np.unique(tags)
plt.bar(tags, values, width=0.3)
```
